It's clear from the above result; that we have to convert n_enrolled, course_effort & course_length cols into integer data types.

```python
# Fill missing values with 0, remove commas, and convert to integers
df['n_enrolled'] = df['n_enrolled'].fillna('0').str.replace(',', '', regex=True).astype(int)

# Assert that the conversion occurred
assert df['n_enrolled'].dtype == 'int'
```

```python
# remove 'weeks' from course_length column, and convert to integers
df['course_length'] = df['course_length'].str.strip('Weeks').astype(int)
#assert that conversion occured
assert df['course_length'].dtype == 'int'
```

```python

import re
import numpy as np

# Sample DataFrame with a 'course_effort' column
data = {'course_effort': ["2-3 hours per week", "4-6 hours per week", "1-2 hours per week", "7-8 hours per week", np.nan, "9-10 hours per week"]}
df1 = pd.DataFrame(data)

# Define a function to extract and convert durations
def extract_and_convert_duration(duration_string):
    if pd.notna(duration_string):  # Check for non-missing values
        match = re.search(r'(\d+)-(\d+)\s*hours per week', duration_string)
        if match:
            min_weeks = int(match.group(1))
            max_weeks = int(match.group(2))
            # Choose either min_weeks or max_weeks depending on your preference
            return min_weeks
    return -1  # Default value for missing or unmatched values

# Apply the function to the 'effort' column
df1['mod_course_effort'] = df1['course_effort'].apply(extract_and_convert_duration)

# Print the updated DataFrame
df1.head(20)
```

```python
new_df = pd.concat([df, df1], axis = 1)
new_df.head(2)
```

```python
#drop useless cols; course_effort
new_df.drop(columns = 'course_effort', inplace = True)
new_df.head(3)
```
